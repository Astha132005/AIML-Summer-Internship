```python
#importing Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#line plot
x=[20,30,45,67,78,39]
y=[30,25,35,16,38,34]
plt.plot(x,y,color='purple')#coordinates ko bas join krega # color='k' mein black color aega
plt.title('Line Plot') # Title dene ke 
plt.xlabel('X-Axis')# koi bhi data ko X-axis mein represent krte waqt agar kuchh defining term dena hai ex:'Age','X-Axis'
plt.ylabel('Y-Axis')
plt.grid()#gives a grid for better representation
plt.show()# above all info has to be brought in 1 plot using show() function
```

```python
#line plot
x=[20,30,45,67,78,39]
y=[30,25,35,16,38,34]
plt.plot(x,y,color='purple')#coordinates ko bas join krega # color='k' mein black color aega
plt.title('Line Plot') # Title dene ke 
plt.show()

# Agar hum yaha move krdete hain toh uske upr ka sara info ek plot mei aega aur uske neeche wala dusre mein
plt.xlabel('X-Axis')# koi bhi data ko X-axis mein represent krte waqt agar kuchh defining term dena hai ex:'Age','X-Axis'
plt.ylabel('Y-Axis')
plt.grid()#gives a grid for better representation
```

```python
#line plot
x=[20,30,45,67,78,39]
y=[30,25,35,16,38,34]
plt.plot(x,y,'8-',linewidth=2,markersize=20,color='y')#coordinates ko bas join krega # color='k' mein black color aega
# here 8- represents 8 point star= octagon
# colour b krne  se blue aega
plt.title('Line Plot ❤️') # Title dene ke 
plt.xlabel('X-Axis')# koi bhi data ko X-axis mein represent krte waqt agar kuchh defining term dena hai ex:'Age','X-Axis'
plt.ylabel('Y-Axis')
plt.grid()#gives a grid for better representation
plt.show()# above all info has to be brought in 1 plot using show() function
```

```python
# Pie Charts:
slices=[8,10,4,3,5]
activities=['sleeping','phone','study','eat','play']
col=['gold','b','magenta','red','g']
plt.pie(slices, labels= activities,colors=col,startangle=90,explode=(0,0,0.3,0,0),shadow=True,autopct='%1.2f%%')
plt.title('Pie Chart')
plt.show()
```

```python
#lolipop chart:
plt.stem(x,y,linefmt='green',markerfmt='ro',basefmt="")
# Format string for the vertical lines (stems). 'green' means green lines.
# Format string for the markers. 'ro' means red (r) circles (o).
# 	Format string for the baseline. '' means no line at the bottom.
plt.title("Lolipop Chart")
```

```python
#bubble chart:
sizes =[30, 100, 300, 200, 150, 250, 90]

colors = [1, 2, 3, 4, 5, 6, 7]

plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.6, edgecolors='black')

plt.colorbar(label='Scale')

plt.title("Bubble Chart")
```

```python
labels =['A', 'B', 'C', 'D', 'E']
values= [3, 5, 2, 8, 7]
angles = np.linspace(0, 2*np.pi, len(labels), endpoint =False).tolist()
values +=values[:1]
angles +=angles[:1]
fig, ax =plt.subplots(subplot_kw={'polar': True})
ax.plot(angles, values, color='blue', linewidth=2)
ax.fill(angles, values, color='skyblue', alpha=0.4)
ax.set_xticks(angles[:-1])
ax.set_xticklabels (labels)
plt.title("Radar Chart")
```

```python
#spiral galaxy plot:
theta= np.linspace(0, 6 *np.pi, 200)
r=np.linspace(0.1, 2, 200)
x=r*np.cos(theta)
y=r*np.sin(theta)
plt.figure(figsize=(6, 6))
plt.scatter(x, y, c=theta, cmap='plasma', s=20)
plt.title("Spiral Galaxy Plot")
plt.axis('off')
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [5, 4, 2, 1, 3, 0, 2]
y2 = [1, 2, 3, 4, 5, 6, 7]
plt.bar(x, y, color='green')
plt.bar(x, y2, bottom=y, color='orange')
plt.title("Stacked Bar Chart")
plt.show()
```

```python
x= np.linspace(1, 7, 7)
y=np.random.randint(1, 10, size=7)
plt.figure(figsize=(7, 5))
for i in range(len(x)):
    plt.text(x[i], y[i], "❤️", fontsize=30, ha='center', va='center')
plt.xlim(0, 8)
plt.ylim(0, 12)
plt.title("Emoji Plot❤️")
plt.grid(True)
plt.show()
```

```python
angles =np.linspace(0, 2* np.pi, 100)
radii= np.random.uniform(0.8, 1.2, size=100)
colors =np.random.rand(100)
x=radii*np.cos(angles)
y=radii*np.sin(angles)
plt.figure(figsize=(6, 6))
plt.scatter(x, y, c=colors, cmap='viridis', s=100, alpha=0.8, edgecolors='black')
plt.axis('off')
plt.title("Donut Dot Plot😊")
plt.show()
```
